```python
def resumo_nulos(df, nome_df):
    total = len(df)
    null_counts = df.isna().sum().sort_values(ascending=False)
    null_ratio = (null_counts / total).round(3)
    resumo = (
      pd.DataFrame({

        "n_nulos": null_counts,
        "perc_nulos": null_ratio,
      })
      .query("n_nulos > 0")
    )
    print(f"\n===== {nome_df} =====")
    return resumo

resumo_nulos(dfc_prep, "COMPRAS")
resumo_nulos(dfcli_prep, "CLIENTES")
resumo_nulos(dfp_prep, "PEDIDOS_CD")
```

```python
import numpy as np

#garante que recency é numérico
rfm_enriq["recency"] = pd.to_numeric(rfm_enriq["recency"], errors="coerce")

LIMITE_PERDA_CADASTRO = 180 #6 meses

condicoes = [
    # 1) NÃO tem nome_master → nunca cadastrado em lugar nenhum
    (rfm_enriq["nome_master"].isna()),
    
    # 2) Tem nome_master E tem_cadastro True → cadastrado neste CD
    #    2a) se recency <= 180 → ativo neste CD
    (rfm_enriq["nome_master"].notna()) &
    (rfm_enriq["tem_cadastro"]) &
    (rfm_enriq["recency"] <= LIMITE_PERDA_CADASTRO),

    #    2b) se recency > 180 → perdeu cadastro neste CD
    (rfm_enriq["nome_master"].notna()) &
    (rfm_enriq["tem_cadastro"]) &
    (rfm_enriq["recency"] > LIMITE_PERDA_CADASTRO),

    # 3) Tem nome_master E tem_cadastro False → cadastrado em outro CD
    #    3a) recency <= 180 → ativo em outro CD
    (rfm_enriq["nome_master"].notna()) &
    (~rfm_enriq["tem_cadastro"]) &
    (rfm_enriq["recency"] <= LIMITE_PERDA_CADASTRO),

    #    3b) recency > 180 → cadastro perdido em outro CD
    (rfm_enriq["nome_master"].notna()) &
    (~rfm_enriq["tem_cadastro"]) &
    (rfm_enriq["recency"] > LIMITE_PERDA_CADASTRO),
]

valores = [
    "Nunca cadastrado",
    "Cadastrado ativo neste CD",
    "Cadastro perdido neste CD (>6m)",
    "Cadastrado ativo em outro CD",
    "Cadastro perdido em outro CD (>6m)",

]

rfm_enriq["status_cadastro"] = np.select(condicoes, valores, default="Indefinido")
```

```python
import numpy as np

# para garantir que recency é numérica
rfm_enriq["recency"] = pd.to_numeric(rfm_enriq["recency"], errors="coerce")

# regra> tem_cadastro = True -> cliente deste CD / False -> cliente de outro CD
rfm_enriq["relacao_cd"] = np.where(
    rfm_enriq["tem_cadastro"],
    "Cliente deste CD",
    "Cliente de outro CD",
)

# Classificar por tempo:
LIMITE_PERDA_CADASTRO = 180

rfm_enriq["situacao_tempo"] = np.where(
    rfm_enriq["recency"] <= LIMITE_PERDA_CADASTRO,
    "Ativo",
    "Inativo",
)

# status combinado (para visualização)
rfm_enriq["status_cadastro"] = (
    rfm_enriq["relacao_cd"] + " | " + rfm_enriq["situacao_tempo"]
)

rfm_enriq[["id_cliente", "nome_cliente", "relacao_cd", "situacao_tempo", "monetary", "recency"]].sort_values("monetary", ascending=False).head(10)
```

```python
resumo_relacao_cd=(
    rfm_enriq
    .groupby("relacao_cd")
    .agg(
        n_clientes=("id_cliente", "nunique"),
        faturamento=("monetary", "sum"),
        freq_media=("frequency", "mean"),
        recency_media=("recency", "mean"),
    )
    .sort_values("faturamento", ascending=False)
)

resumo_relacao_cd
```

```python
# Fazer a segmentação RFM (VIP, Em risco, etc.)

#cria scores de 1 a 5 por quantil
rfm_enriq["R_score"] = pd.qcut(rfm_enriq["recency"], 5, labels=[5,4,3,2,1])
rfm_enriq["F_score"] = pd.qcut(rfm_enriq["frequency"].rank(method="first"), 5, labels=[1,2,3,4,5])
rfm_enriq["M_score"] = pd.qcut(rfm_enriq["monetary"].rank(method="first"), 5, labels=[1,2,3,4,5])

#descartar só as linhas com algum NaN em R/F/M
mask_ruim = rfm_enriq[["R_score", "F_score", "M_score"]].isna().any(axis=1)


# Mantém quem tem score completo
rfm_scored = rfm_enriq.loc[~mask_ruim].copy()

# converte os scores para int
rfm_scored["R_score"] = rfm_scored["R_score"].astype(int)
rfm_scored["F_score"] = rfm_scored["F_score"].astype(int)
rfm_scored["M_score"] = rfm_scored["M_score"].astype(int)


rfm_scored["RFM_score"] = rfm_scored["R_score"]*100 + rfm_scored["F_score"]*10 + rfm_scored["M_score"]
```

```python
def rotula_segmento(row):
    if row["R_score"] >=4 and row["F_score"] >=4 and row["M_score"] >=4:
        return "VIP"
    if row["R_score"] >=4 and row["F_score"] >=3:
        return "Frequentes recentes"
    if row["R_score"] <=2 and row["F_score"] >=4:
        return "Em risco"
    if row["F_score"] <=2 and row["F_score"] <=2:
        return "Baixo valor"
    return "Intermediário"

rfm_scored["segmento_rfm"] = rfm_scored.apply(rotula_segmento, axis=1)

rfm_scored["segmento_rfm"].value_counts()
```

```python
regioes = (
    rfm_geo_valid
    .groupby("regiao_id")
    .agg(
        clientes_total=("id_cliente", "nunique"),
        cliente_cd=("relacao_cd", lambda s: (s == "Cliente deste CD").sum()
                    if "relacao_cd" in rfm_geo_valid.columns else s.size),
        clientes_outros_cd=("relacao_cd", lambda s: (s == "Clietne de outro CD").sum()
                            if "relacao_cd" in rfm_geo_valid.columns else 0),
        vip_total=("segmento_rfm", lambda s: (s =="VIP").sum()
                   if "segmento_rfm" in rfm_geo_valid.columns else 0),
        faturamento_total=("monetary", "sum"),
        lat=("lat", "mean"),
        lon=("lon", "mean"),
    )
    .reset_index()
)

regioes.sort_values("faturamento_total", ascending=False).head(10)
```

```python
import folium

# se não houver nenhuma região, só pra garantir
if regioes.empty:
    print("Nenhuma região com coordenadas.")
else:
    centro_lat = regioes["lat"].mean()
    centro_lon = regioes["lon"].mean()

    m_regioes = folium.Map(location=[centro_lat, centro_lon], zoom_start=12)

# normaliza o faturamento para definir tamanho das bolhas
    max_fat = regioes["faturamento_total"].max()
    if max_fat == 0:
        max_fat = 1 #evita divisão por zero

    for _, row in regioes.iterrows():
        peso = row["faturamento_total"] / max_fat # entre 0 e 1
        radius = 100 + 500 * peso # ajusta o tamanho das bolhas

        popup_text = (
          f"Região: {row['regiao_id']}<br>"
            f"Clientes total: {row['clientes_total']}<br>"
            f"VIPs: {row['vip_total']}<br>"
            f"Clientes deste CD: {row['cliente_cd']}<br>"
            f"Clientes outro CD: {row['clientes_outros_cd']}<br>"
            f"Faturamento (período): R$ {row['faturamento_total']:.2f}"
        )

        folium.Circle(
            location=[row["lat"], row["lon"]],
            radius=radius,
            popup=popup_text,
            fill=True,
            fill_opacity=0.6,
        ).add_to(m_regioes)

    m_regioes
```
